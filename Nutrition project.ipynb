```python
#Renaming the row values of Gender column using replace()
```

```python
df_malnutrition['Gender'] = df_malnutrition['Gender'].replace({'SEX_FMLE': 'Female', 'SEX_BTSX': 'Both', 'SEX_MLE': 'Male'})
```

```python
# Changing the names of Countries that are given as three letter codes in the country column to their full names using 'pycountry'
```

```python
!pip install pycountry
```

```python
import pycountry
special_cases = {
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region',
    'WB_UMI': 'Upper Middle Income'
}
def get_country_full_name(code):
    if code in special_cases:
        return special_cases[code]
    country = pycountry.countries.get(alpha_3=code)
    if country:
        return country.name
    else:
        return None
df_malnutrition['Country_name'] = df_malnutrition['Country'].apply(get_country_full_name)
df_obesity['Country_name'] = df_obesity['Country'].apply(get_country_full_name)
```

```python
df_malnutrition
```

```python
df_obesity
```

```python
#adding new column CI_Width( confidence interval width-range within which a population parameter is estimated to lie, with a certain level of confidence)
```

```python
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']
```

```python
df_obesity
```

```python
df_malnutrition['CI_Width'] = df_malnutrition['UpperBound'] - df_malnutrition['LowerBound']
```

```python
df_malnutrition
```

```python
# Categorizing obesity based on Mean estimate values as 'High','Moderate'and 'Low' by defining a function 
# and adding a column obesity_level to obesity dataframe
```

```python
def categorize_obesity(Mean_Estimate):
    if Mean_Estimate >= 30:
        return 'High'
    elif 25 <= Mean_Estimate < 30:
        return 'Moderate'
    else:
        return 'Low' 
df_obesity['obesity_level'] = df_obesity['Mean_Estimate'].apply(categorize_obesity)
```

```python
df_obesity
```

```python
# Categorizing malnutrition based on Mean estimate values as 'High','Moderate'and 'Low' by defining a function 
# and adding a column malnutrition_level to malnutrition dataframe
```

```python
def categorize_malnutrition(Mean_Estimate):
    if Mean_Estimate >= 30:
        return 'High'
    elif 25 <= Mean_Estimate < 30:
        return 'Moderate'
    else:
        return 'Low' 
df_malnutrition['malnutrition_level'] = df_malnutrition['Mean_Estimate'].apply(categorize_malnutrition)
```

```python
df_malnutrition
```

```python
#deleting country column
```

```python
df_obesity.drop(['Country'],axis = 1, inplace = True)
df_malnutrition.drop(['Country'],axis = 1, inplace = True)
```

```python
df_obesity = df_obesity.rename(columns={'Country_name': 'Country'}) #renaming Country_name
df_malnutrition = df_malnutrition.rename(columns={'Country_name': 'Country'})
```

```python
import pymysql
connection = pymysql.connect(host = "localhost",user = "root",password = "password",database = "nutrition") #connecting with database
cursor = connection.cursor()
```

## Obesity_Table

```python
#creating table
cursor.execute("""
            create table obesity(
            Region text,
            Gender varchar(20),
            Year int,
            Mean_Estimate float,
            LowerBound float,
            UpperBound float,
            age_group varchar(20),
            Country text,
            CI_Width float,
            obesity_level varchar(20));
            """ )
```

```python
#inserting values into table
for index, row in df_obesity.iterrows():
    cursor.execute("""
        INSERT INTO obesity (
            Region, Gender, Year, Mean_Estimate, LowerBound, UpperBound, 
            age_group, Country, CI_Width, obesity_level
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Region'], row['Gender'], row['Year'], row['Mean_Estimate'],
        row['LowerBound'], row['UpperBound'], row['age_group'],
        row['Country'], row['CI_Width'], row['obesity_level']
    ))
connection.commit()
```

## Malnutrition_Table

```python
#creating table
cursor.execute("""
            create table malnutrition(
            Region text,
            Gender varchar(20),
            Year int,
            Mean_Estimate float,
            LowerBound float,
            UpperBound float,
            age_group varchar(20),
            Country text,
            CI_Width float,
            malnutrition_level varchar(20));
            """ )
```

```python
#inserting values into table
for index, row in df_malnutrition.iterrows():
    cursor.execute("""
        INSERT INTO malnutrition (
            Region, Gender, Year, Mean_Estimate, LowerBound, UpperBound, 
            age_group, Country, CI_Width, malnutrition_level
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Region'], row['Gender'], row['Year'], row['Mean_Estimate'],
        row['LowerBound'], row['UpperBound'], row['age_group'],
        row['Country'], row['CI_Width'], row['malnutrition_level']
    ))
connection.commit()
```
